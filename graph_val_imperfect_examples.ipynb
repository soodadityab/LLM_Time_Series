```python
import pandas as pd
import matplotlib.pyplot as plt

# Simulate the data
data = {
    'data_dte': ['2021-01-01', '2021-02-01', '2021-03-01', '2021-04-01', '2021-05-01'],
    'Year': [2021, 2021, 2021, 2021, 2021],
    'Month': [1, 2, 3, 4, 5],
    'usg_apt_id': [1, 2, 3, 4, 5],
    'usg_apt': ['JFK', 'LAX', 'ORD', 'DFW', 'DEN'],
    'usg_wac': [10, 20, 30, 40, 50],
    'fg_apt_id': [1, 2, 3, 4, 5],
    'fg_apt': ['ATL', 'LAX', 'ORD', 'DFW', 'DEN'],
    'fg_wac': [10, 20, 30, 40, 50],
    'airlineid': [100, 101, 102, 103, 104],
    'carrier': ['AA', 'DL', 'UA', 'SW', 'B6'],
    'carriergroup': [1, 1, 1, 1, 1],
    'type': ['Domestic', 'International', 'Domestic', 'Domestic', 'International'],
    'Scheduled': [1000, 900, 1100, 950, 980],
    'Charter': [100, 120, 110, 105, 100],
    'Total': [1100, 1020, 1210, 1055, 1080]
}

df = pd.DataFrame(data)

# Group data by year
grouped_df = df.groupby('Year').sum()

# Line plot of scheduled and chartered flights per year
plt.figure(figsize=(10, 6))
plt.plot(grouped_df.index, grouped_df['Scheduled'], label='Scheduled', marker='o')
plt.plot(grouped_df.index, grouped_df['Charter'], label='Charter', marker='o')
plt.xlabel('Year')
plt.ylabel('Number of Flights')
plt.title('Scheduled and Chartered Flights per Year')
plt.legend()
plt.show()
#score: 2/3
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Simulate the data
data = {
    'Timestamp': ['2021-01-01', '2021-02-01', '2021-03-01', '2021-04-01', '2021-05-01'],
    'Age': [25, 30, 35, 40, 45],
    'Gender': ['Male', 'Female', 'Male', 'Female', 'Male'],
    'City': ['Berlin', 'Munich', 'Hamburg', 'Frankfurt', 'Stuttgart'],
    'Position': ['Developer', 'Manager', 'Developer', 'Consultant', 'Developer'],
    'Total years of experience': [3, 5, 7, 10, 15],
    'Years of experience in Germany': [2, 4, 5, 7, 10],
    'Seniority level': ['Junior', 'Mid', 'Senior', 'Lead', 'Senior'],
    'Your main technology / programming language': ['Python', 'Java', 'C++', 'JavaScript', 'Python'],
    'Other technologies/programming languages you use often': ['SQL', 'SQL', 'SQL', 'SQL', 'SQL'],
    'Yearly brutto salary (without bonus and stocks) in EUR': [50000, 60000, 70000, 80000, 90000],
    'Number of vacation days': [25, 26, 27, 28, 30],
    'Employment status': ['Employed', 'Employed', 'Employed', 'Employed', 'Employed'],
    'ontract duration': ['Permanent', 'Permanent', 'Permanent', 'Permanent', 'Permanent'],
    'Main language at work': ['English', 'German', 'English', 'German', 'English'],
    'Company size': ['Small', 'Medium', 'Large', 'Large', 'Small'],
    'Company type': ['Startup', 'Enterprise', 'Consulting', 'Consulting', 'Startup'],
    'Have you lost your job due to the coronavirus outbreak?': ['No', 'No', 'No', 'No', 'No']
}

df = pd.DataFrame(data)

# KDE plot of the distribution of total years of experience
plt.figure(figsize=(10, 6))
sns.kdeplot(df['Total years of experience'], shade=True)
plt.xlabel('Total Years of Experience')
plt.title('Distribution of Total Years of Experience')
plt.show()

#score: 1/3
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Simulate the data (same as above)

df = pd.DataFrame(data)

# Scatter plot of age vs total years of experience, colored by gender
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='Age', y='Total years of experience', hue='Gender')
plt.xlabel('Age')
plt.ylabel('Total Years of Experience')
plt.title('Age vs Total Years of Experience')
plt.show()

#score: 3/3
```
